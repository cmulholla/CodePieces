```python
from sklearn import datasets
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
from mpl_toolkits import mplot3d
%matplotlib inline
```

```python
data1 = pd.read_csv("cities/worldcities2.csv") 
data1
```

```python
plt.scatter(data1['lng'], data1['lat'])
plt.xlim(-200,200)
plt.ylim(-60,80)
plt.show()
```

```python
plt.clf()
x = data1.iloc[:,2:4] # 1t for rows and second for columns
x
```

```python
k=6
kmeans = KMeans(k)
kmeans.fit(x)
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = data1.copy()
data_with_clusters['Clusters'] = identified_clusters 
```

```python
plt.scatter(data_with_clusters['lng'],data_with_clusters['lat'],c=data_with_clusters['Clusters'],cmap='rainbow')
plt.savefig('world'+str(k)+'.png', format="png")
```
